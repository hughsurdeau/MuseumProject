```python
def get_mean_wanderer_ratio(dataframe):
    #Returns #wanderers/(wanderers + walkers)
    walkers = dataframe['linear walkers'].tolist()
    wanderers = dataframe['wanderers'].tolist()
    total = [x+y for x,y in zip(walkers, wanderers)]
    return sum(wanderers) / sum(total)


def get_mean_agents(dataframe):
    walkers = dataframe['linear walkers'].tolist()
    wanderers = dataframe['wanderers'].tolist()
    total = [x+y for x,y in zip(walkers, wanderers)]
    return statistics.mean(total)
    
```
