```python
#AT10
data_file = "/Users/hughsurdeau/PycharmProjects/MuseumProject/data/csv/linear_museum_10AT.csv"
df = pd.read_csv(data_file)


mycolors = ['peru', 'thistle']

fig, ax = plt.subplots(1,1,figsize=(16, 9), dpi= 80)
columns = df.columns[1:3]
labs = columns.values.tolist()

x = df['t'].values.tolist()
flowers = df['linear walkers'].values.tolist()
wanderers = df['wanderers'].values.tolist()

y = np.vstack([flowers, wanderers])

# Plot for each column
labs = columns.values.tolist()
ax = plt.gca()
ax.stackplot(x, y, labels=labs, colors=mycolors, alpha=0.8)

# Decorations
ax.set_title('Wandering vs Linear Visitors in Linear Museum (0.1 Wandering Chance)', fontsize=18)
#ax.set(ylim=[0, 100])
ax.legend(fontsize=10, ncol=4)
ax.set_xlabel("Time")
ax.set_ylabel("Number of Visitors Adhering to Norm")
plt.xticks(x[::100], fontsize=10, horizontalalignment='center')
#plt.yticks(np.arange(0, 101, 20), fontsize=10)
plt.xlim(x[0], x[-1])

# Lighten borders
plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.3)

plt.savefig("/Users/hughsurdeau/Desktop/Bath/Dissertation/Images/LinearModelGraphs/AssholeThreshold/WanderersVsFlowers/linear10AT.png")
```
